```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data in array format
arr = np.array([
    [2104, 400],
    [1600, 330],
    [2400, 369],
    [1416, 232],
    [3000, 540]
])
```

```python
plt.scatter(arr[:, 0], arr[:, 1])
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
# Data in DataFrame format
df = pd.DataFrame(arr, columns=["Area", "Price"])
df
```

```python
# Separate the feature matrix and the label vector
x = arr[:, 0].reshape((-1, 1))
y = arr[:, 1]
x.shape, y.shape
```

```python
def h(x,t):
  """
  Hypothesis function
  """
  return x@t
```

```python
# learning rate
a = 1e-8
# Iteration Count
it = 0
# Number of samples
m = 5
# Number of features
n = 1
# Initialize parameters with zero
th = np.zeros(n, dtype='float')
# Plot line after every 5 iterations
plot_iteration = 5
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(7)
plt.scatter(x[:,0],y)

# Initial loss
J = (1/(2*m))*((h(x,th)-y)@(h(x,th)-y).T)
while (it<40):
  th -= a*(x@th - y).T@x/m
  k = J
  J = (1/(2*m))*((x@th-y)@(x@th-y).T)
  it += 1
  if it%plot_iteration == 0:
    plt.plot(x[:,0],h(x,th),label=f'{it}th iteration')
  if abs(k-J)<1e-9:
        break
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.show()
```
